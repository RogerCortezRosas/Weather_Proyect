```python
import pandas as pd
import re
import pymysql
import pandas as pd
from io import BytesIO
from sqlalchemy import create_engine
import numpy as np
```

```python

def load_to_bd(df,name):
    # Datos de conexiÃ³n a la RDS de Amazon
    host = '127.0.0.1'  # Cambia esto por el endpoint de tu RDS
    port = 3307
    user = 'airflow'  # Cambia esto por tu usuario
    password = 'airflow'  # Cambia esto por tu contraseÃ±a
    database = 'airflow'  # Cambia esto por tu base de datos
    engine = None

    try:
        engine = create_engine(f'mysql+pymysql://{user}:{password}@{host}:{port}/{database}')

        df.to_sql(name,con=engine,if_exists='append',index=False)
        return print("ConexiÃ³n y insercion exitosa a la base de datos RDS MySQL")

        

    

    except pymysql.MySQLError as e:
        print(f"Error al conectar a la base de datos: {e}")
    finally:
        if engine:
            engine.dispose()
```
