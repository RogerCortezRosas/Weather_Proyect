```python
import pandas as pd
import re
import pymysql
import pandas as pd
from io import BytesIO
from sqlalchemy import create_engine
import numpy as np
```

```python

def load_to_bd(df,name):
    # Datos de conexión a la RDS de Amazon
    host = '127.0.0.1'  # Cambia esto por el endpoint de tu RDS
    port = 3306
    user = 'root'  # Cambia esto por tu usuario
    password = 'root1234'  # Cambia esto por tu contraseña
    database = 'weather'  # Cambia esto por tu base de datos
    engine = None

    try:
        engine = create_engine(f'mysql+pymysql://{user}:{password}@{host}:{port}/{database}')

        df.to_sql(name,con=engine,if_exists='append',index=False)
        return print("Conexión y insercion exitosa a la base de datos RDS MySQL")

        

    

    except pymysql.MySQLError as e:
        print(f"Error al conectar a la base de datos: {e}")
    finally:
        if engine:
            engine.dispose()
```

```python
dicc = {
    'humidity': humidity_df,
    'pressure': pressure_df,
    'temperature': temperature_df,
    'weather_description': weatherDes_df,
    'wind_direction': windDir_df,
    'wind_speed': wind_speed_df
}
```

```python
for key,df in dicc.items():
    df = df.iloc[0:15, :]
    load_to_bd(df,key)
```
